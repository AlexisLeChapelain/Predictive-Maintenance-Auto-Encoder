```python
# define a cursor for the connection
cur = conn.cursor()

# SQL query to get the public tables
query = """
SELECT
    table_schema || '.' || table_name
FROM
    information_schema.tables
WHERE
    table_type = 'BASE TABLE'
AND
    table_schema NOT IN ('pg_catalog', 'information_schema');
"""

# run the query
cur.execute(query)

# store the results of the query in a python variable
tables = cur.fetchall() 
```

```python
for n in tables:
    print(n)
```
